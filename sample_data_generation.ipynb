```python
import csv
from IPython.display import FileLink, display

# Define a list of country names
country_names = [
    "SOUTH SUDAN",
    "DEMOCRATIC REPUBLIC of CONGO",
    "PALESTAN",
    "YEMEN",
    "LEBANON",
    "CAMBODIA",
    "KENYA",
    "SYRIAN ARAB REPUBLIC",
    "SOMALIA",
    "IRAQ",
    "ETHIOPIA",
    "GUATEMALA",
    "BURKINA FASO",
    "PHILIPPINES",
    "UKRAINE",
    "BANGLADESH",
    "UGANDA",
    "CENTRAL AFRICAN REPUBLIC",
    "MYANMAR",
    "JORDAN"
]

# Function to generate a code: first three letters of the string (remove spaces, uppercase)
def get_code(name):
    # Remove spaces and take the first 3 letters, then convert to uppercase
    code = ''.join(name.split())[:3].upper()
    return code

# Generate a list of country dictionaries with country_ID and country_name.
# Here, country_ID is generated as the first three letters of the country name.
countries = []
for name in country_names:
    country = {
        "country_id": get_code(name),
        "country_name": name
    }
    countries.append(country)

# Initialize a list to store region data
regions = []

# For each country, generate 4 regions: two urban and two rural.
# The region_ID is generated by combining the country's country_ID with "01", "02", "03", "04".
for country in countries:
    country_id = country["country_id"]
    country_name = country["country_name"]
    for i in range(1, 5):
        if i <= 2:
            region_type = "urban"
            region_label = f"Urban {i}"
        else:
            region_type = "rural"
            region_label = f"Rural {i-2}"

        # Build region_ID: country_id + two-digit number
        region_id = f"{country_id}{i:02d}"
        # Build region_name: e.g., "SOUTH SUDAN Urban 1"
        region_name = f"{country_name} {region_label}"

        regions.append({
            "region_id": region_id,
            "region_name": region_name,
            "region_type": region_type
        })

# Write country data to a CSV file for MySQL import.
with open('countries.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['country_id', 'country_name']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for country in countries:
        writer.writerow(country)

# Write region data to a CSV file for MySQL import.
with open('regions.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['region_id', 'region_name', 'region_type']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for region in regions:
        writer.writerow(region)

print("CSV files 'countries.csv' and 'regions.csv' have been generated.")

# Display links to download the files (if running in a Jupyter Notebook)
display(FileLink("countries.csv"))
display(FileLink("regions.csv"))
```

```python
import random
import csv
from datetime import datetime, timedelta

# Function to generate a random date between two dates
def random_date(start, end):
    """
    Generate a random date between start and end.
    """
    delta = end - start
    random_days = random.randint(0, delta.days)
    return start + timedelta(days=random_days)

# Define the list of years for project generation (2015 to 2024)
years = list(range(2015, 2025))

# Define the list of SDG indicators (equal weight here, can adjust if needed)
sdg_list = ["SDG_1.1.1", "SDG_5.4.1", "SDG_13.1.1"]

# Define the four project categories and their weight distributions based on region type.
# For urban regions: Education (35%), Advertising Campaign (30%), Financial Aid (20%), Infrastructure (15%)
urban_categories = ["Education", "Advertising Campaign", "Financial Aid", "Infrastructure"]
urban_weights = [35, 30, 20, 15]

# For rural regions: Financial Aid (35%), Infrastructure (30%), Education (20%), Advertising Campaign (15%)
rural_categories = ["Financial Aid", "Infrastructure", "Education", "Advertising Campaign"]
rural_weights = [35, 30, 20, 15]

# Define project count range for each region-year cell.
# We want each region-year to have at least 1 project. Here, we use:
# Urban: between 7 and 9 projects per region-year
# Rural: between 8 and 11 projects per region-year
def get_project_count(region_type):
    if region_type == "urban":
        return random.randint(7, 9)
    else:  # rural
        return random.randint(8, 11)

# Define budget range based on region type:
# Urban: 400,000 to 800,000
# Rural: 600,000 to 1,000,000
def get_budget(region_type):
    if region_type == "urban":
        return random.randint(400000, 800000)
    else:
        return random.randint(600000, 1000000)

# Function to generate a random start_date within a given year
def generate_start_date(year):
    start = datetime(year, 1, 1)
    end = datetime(year, 12, 31)
    return random_date(start, end)

# Function to generate a random end_date given a start_date and maximum end year (2024)
def generate_end_date(start_date, max_year=2024):
    start_year = start_date.year
    # Choose a random end year between start_year and max_year
    end_year = random.randint(start_year, max_year)
    # If end year is the same as start year, ensure end_date is not before start_date
    if end_year == start_year:
        end = datetime(end_year, 12, 31)
        # Generate a date between start_date and December 31 of start_year
        return random_date(start_date, end)
    else:
        # Generate a random date in the chosen end_year
        start_of_end_year = datetime(end_year, 1, 1)
        end_of_end_year = datetime(end_year, 12, 31)
        return random_date(start_of_end_year, end_of_end_year)

# Assume 'regions' is already generated from the previous step.
# For demonstration, here's a dummy 'regions' list with few sample records.
# In practice, this list should contain all 80 regions generated earlier.
# Example format: {"region_id": "SOU01", "region_name": "SOUTH SUDAN Urban 1", "region_type": "urban", "country_id": "SOU"}
# Uncomment below if you want to test with a small sample:
# regions = [
#     {"region_id": "SOU01", "region_name": "SOUTH SUDAN Urban 1", "region_type": "urban", "country_id": "SOU"},
#     {"region_id": "SOU02", "region_name": "SOUTH SUDAN Urban 2", "region_type": "urban", "country_id": "SOU"},
#     {"region_id": "SOU03", "region_name": "SOUTH SUDAN Rural 1", "region_type": "rural", "country_id": "SOU"},
#     {"region_id": "SOU04", "region_name": "SOUTH SUDAN Rural 2", "region_type": "rural", "country_id": "SOU"}
# ]
# In actual use, 'regions' should contain all 80 regions from step 1.

# Initialize list to store all generated project records
projects = []
project_id = 1

# Loop through each region and each year to generate projects
for region in regions:
    region_id = region["region_id"]
    region_type = region["region_type"]
    # For each year from 2015 to 2024
    for year in years:
        # Ensure each region-year has at least one project; here we use a random range based on region type
        count = get_project_count(region_type)
        for _ in range(count):
            # Generate project attributes
            proj_id = project_id
            proj_name = f"Project_{proj_id}"

            # Choose project category based on region type weight distribution
            if region_type == "urban":
                category = random.choices(urban_categories, weights=urban_weights, k=1)[0]
            else:
                category = random.choices(rural_categories, weights=rural_weights, k=1)[0]

            # Choose SDG indicator (equal probability)
            sdg = random.choice(sdg_list)

            # Generate start_date within the current year
            start_date_obj = generate_start_date(year)
            start_date = start_date_obj.strftime("%Y-%m-%d")

            # Generate end_date based on start_date
            end_date_obj = generate_end_date(start_date_obj, max_year=2024)
            end_date = end_date_obj.strftime("%Y-%m-%d")

            # Generate budget based on region type
            budget = get_budget(region_type)

            # Create the project record
            project = {
                "project_id": proj_id,
                "project_name": proj_name,
                "category": category,
                "track_sdg_indicator": sdg,
                "start_date": start_date,
                "end_date": end_date,
                "budget": budget,
                "region_id": region_id
            }

            projects.append(project)
            project_id += 1

# Write the generated project data to a CSV file for MySQL import
with open('projects.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ["project_id", "project_name", "category", "track_sdg_indicator", "start_date", "end_date", "budget", "region_id"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for project in projects:
        writer.writerow(project)

print("Project data generated and saved to 'projects.csv'.")

# (Optional) Compute and display total annual budget for each year for verification
annual_budget = {}
for year in years:
    annual_budget[year] = sum(project["budget"] for project in projects if int(project["start_date"][:4]) == year)

print("Annual Budget Totals:")
for year, total in annual_budget.items():
    print(f"{year}: £{total:,}")
import random
import csv
from datetime import datetime, timedelta

# Function to generate a random date between two dates
def random_date(start, end):
    """
    Generate a random date between start and end.
    """
    delta = end - start
    random_days = random.randint(0, delta.days)
    return start + timedelta(days=random_days)

# Define the list of years for project generation (2015 to 2024)
years = list(range(2015, 2025))

# Define the list of SDG indicators (equal weight here, can adjust if needed)
sdg_list = ["SDG_1.1.1", "SDG_5.4.1", "SDG_13.1.1"]

# Define the four project categories and their weight distributions based on region type.
# For urban regions: Education (35%), Advertising Campaign (30%), Financial Aid (20%), Infrastructure (15%)
urban_categories = ["Education", "Advertising Campaign", "Financial Aid", "Infrastructure"]
urban_weights = [35, 30, 20, 15]

# For rural regions: Financial Aid (35%), Infrastructure (30%), Education (20%), Advertising Campaign (15%)
rural_categories = ["Financial Aid", "Infrastructure", "Education", "Advertising Campaign"]
rural_weights = [35, 30, 20, 15]

# Define project count range for each region-year cell.
# We want each region-year to have at least 1 project. Here, we use:
# Urban: between 7 and 9 projects per region-year
# Rural: between 8 and 11 projects per region-year
def get_project_count(region_type):
    if region_type == "urban":
        return random.randint(7, 9)
    else:  # rural
        return random.randint(8, 11)

# Define budget range based on region type:
# Urban: 400,000 to 800,000
# Rural: 600,000 to 1,000,000
def get_budget(region_type):
    if region_type == "urban":
        return random.randint(400000, 800000)
    else:
        return random.randint(600000, 1000000)

# Function to generate a random start_date within a given year
def generate_start_date(year):
    start = datetime(year, 1, 1)
    end = datetime(year, 12, 31)
    return random_date(start, end)

# Function to generate a random end_date given a start_date and maximum end year (2024)
def generate_end_date(start_date, max_year=2024):
    start_year = start_date.year
    # Choose a random end year between start_year and max_year
    end_year = random.randint(start_year, max_year)
    # If end year is the same as start year, ensure end_date is not before start_date
    if end_year == start_year:
        end = datetime(end_year, 12, 31)
        # Generate a date between start_date and December 31 of start_year
        return random_date(start_date, end)
    else:
        # Generate a random date in the chosen end_year
        start_of_end_year = datetime(end_year, 1, 1)
        end_of_end_year = datetime(end_year, 12, 31)
        return random_date(start_of_end_year, end_of_end_year)

# Assume 'regions' is already generated from the previous step.
# For demonstration, here's a dummy 'regions' list with few sample records.
# In practice, this list should contain all 80 regions generated earlier.
# Example format: {"region_id": "SOU01", "region_name": "SOUTH SUDAN Urban 1", "region_type": "urban", "country_id": "SOU"}
# Uncomment below if you want to test with a small sample:
# regions = [
#     {"region_id": "SOU01", "region_name": "SOUTH SUDAN Urban 1", "region_type": "urban", "country_id": "SOU"},
#     {"region_id": "SOU02", "region_name": "SOUTH SUDAN Urban 2", "region_type": "urban", "country_id": "SOU"},
#     {"region_id": "SOU03", "region_name": "SOUTH SUDAN Rural 1", "region_type": "rural", "country_id": "SOU"},
#     {"region_id": "SOU04", "region_name": "SOUTH SUDAN Rural 2", "region_type": "rural", "country_id": "SOU"}
# ]
# In actual use, 'regions' should contain all 80 regions from step 1.

# Initialize list to store all generated project records
projects = []
project_id = 1

# Loop through each region and each year to generate projects
for region in regions:
    region_id = region["region_id"]
    region_type = region["region_type"]
    # For each year from 2015 to 2024
    for year in years:
        # Ensure each region-year has at least one project; here we use a random range based on region type
        count = get_project_count(region_type)
        for _ in range(count):
            # Generate project attributes
            proj_id = project_id
            proj_name = f"Project_{proj_id}"

            # Choose project category based on region type weight distribution
            if region_type == "urban":
                category = random.choices(urban_categories, weights=urban_weights, k=1)[0]
            else:
                category = random.choices(rural_categories, weights=rural_weights, k=1)[0]

            # Choose SDG indicator (equal probability)
            sdg = random.choice(sdg_list)

            # Generate start_date within the current year
            start_date_obj = generate_start_date(year)
            start_date = start_date_obj.strftime("%Y-%m-%d")

            # Generate end_date based on start_date
            end_date_obj = generate_end_date(start_date_obj, max_year=2024)
            end_date = end_date_obj.strftime("%Y-%m-%d")

            # Generate budget based on region type
            budget = get_budget(region_type)

            # Create the project record
            project = {
                "project_id": proj_id,
                "project_name": proj_name,
                "category": category,
                "track_sdg_indicator": sdg,
                "start_date": start_date,
                "end_date": end_date,
                "budget": budget,
                "region_id": region_id
            }

            projects.append(project)
            project_id += 1

# Write the generated project data to a CSV file for MySQL import
with open('projects.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ["project_id", "project_name", "category", "track_sdg_indicator", "start_date", "end_date", "budget", "region_id"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for project in projects:
        writer.writerow(project)

print("Project data generated and saved to 'projects.csv'.")

# (Optional) Compute and display total annual budget for each year for verification
annual_budget = {}
for year in years:
    annual_budget[year] = sum(project["budget"] for project in projects if int(project["start_date"][:4]) == year)

print("Annual Budget Totals:")
for year, total in annual_budget.items():
    print(f"{year}: £{total:,}")
```

```python
import csv
import random

# Step 1: Load region data from CSV file
def load_regions(csv_file='regions.csv'):
    """
    Load region data from CSV file.
    Each row must contain at least 'region_id'.
    """
    regions = []
    with open(csv_file, 'r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        for row in reader:
            region = {
                "region_id": row["region_id"]
                # Optionally, include other fields such as 'region_name', 'region_type' if needed.
            }
            regions.append(region)
    return regions

# Step 2: Generate yearly parameters for survey indicators (2015-2024)
def generate_yearly_params():
    """
    Generate parameters for each year (2015 to 2024):
      - literacy_rate: baseline 0.60, increases by 1-2% per year (occasionally drops).
      - numeracy_rate: baseline 0.55, similar trend.
      - poverty_rate: baseline 0.40, decreases by 1-2% per year (occasionally increases).
    """
    params = {}
    params[2015] = {
        "literacy_rate": 0.60,
        "numeracy_rate": 0.55,
        "poverty_rate": 0.40
    }
    for year in range(2016, 2025):
        prev = params[year - 1]
        # Generate literacy_rate change
        if random.random() < 0.1:  # 10% chance for a drop
            delta_lit = -random.uniform(0.01, 0.02)
        else:
            delta_lit = random.uniform(0.01, 0.02)
        new_lit = max(0.0, min(1.0, prev["literacy_rate"] + delta_lit))

        # Generate numeracy_rate change
        if random.random() < 0.1:
            delta_num = -random.uniform(0.01, 0.02)
        else:
            delta_num = random.uniform(0.01, 0.02)
        new_num = max(0.0, min(1.0, prev["numeracy_rate"] + delta_num))

        # Generate poverty_rate change (lower is better)
        if random.random() < 0.1:
            delta_pov = random.uniform(0.01, 0.02)  # occasional worsening
        else:
            delta_pov = -random.uniform(0.01, 0.02)
        new_pov = max(0.0, min(1.0, prev["poverty_rate"] + delta_pov))

        params[year] = {
            "literacy_rate": new_lit,
            "numeracy_rate": new_num,
            "poverty_rate": new_pov
        }
    return params

# Step 3: Generate INDIVIDUAL and SURVEY data for each region-year
def generate_data(regions, yearly_params, sample_size=5000):
    """
    For each region and each year (2015-2024), generate 'sample_size' individuals and corresponding survey records.

    INDIVIDUAL attributes:
      - personal_id: auto-increment integer.
      - sex: randomly chosen from 'male', 'female', 'prefer not to say' (approx. 49%, 49%, 2%).
      - age: random integer between 14 and 80.
      - lives_in_region_id: assigned from the loaded regions.

    SURVEY attributes:
      - survey_id: same as personal_id.
      - proficiency_in_literacy: True with probability equal to that year's literacy_rate.
      - proficiency_in_numeracy: True with probability equal to that year's numeracy_rate.
      - employment_status: if age < 18 then 'below_minimum_legal_work_age', else 70% 'employed', 30% 'unemployed'.
      - below_poverty_line: True with probability equal to that year's poverty_rate.
      - time_spent_on_unpaid_domestic_work: generated based on sex:
            * male: 700 to 1200 hours.
            * female: 1500 to 2500 hours.
            * prefer not to say: randomly choose from male or female range.
      - survey_year: the current year.
      - takes_by_personal_id: same as personal_id.
    """
    individuals = []
    surveys = []
    personal_id = 1

    for region in regions:
        region_id = region["region_id"]
        for year in range(2015, 2025):
            params = yearly_params[year]
            for _ in range(sample_size):
                # Generate sex distribution
                r = random.random()
                if r < 0.49:
                    sex = "male"
                elif r < 0.98:
                    sex = "female"
                else:
                    sex = "prefer not to say"

                # Generate age between 14 and 80
                age = random.randint(14, 80)

                # Create INDIVIDUAL record
                indiv = {
                    "personal_id": personal_id,
                    "sex": sex,
                    "age": age,
                    "lives_in_region_id": region_id
                }
                individuals.append(indiv)

                # Generate SURVEY attributes based on yearly parameters
                proficiency_in_literacy = random.random() < params["literacy_rate"]
                proficiency_in_numeracy = random.random() < params["numeracy_rate"]

                if age < 18:
                    employment_status = "below_minimum_legal_work_age"
                else:
                    employment_status = random.choices(["employed", "unemployed"], weights=[70, 30])[0]

                below_poverty_line = random.random() < params["poverty_rate"]

                # Generate average_time_spent_on_unpaid_domestic_work_weekly based on sex
                if sex == "male":
                    time_spent = random.randint(12, 28)
                elif sex == "female":
                    time_spent = random.randint(23, 50)
                else:  # prefer not to say
                    if random.random() < 0.5:
                        time_spent = random.randint(12, 28)
                    else:
                        time_spent = random.randint(23, 50)

                survey = {
                    "survey_id": personal_id,
                    "proficiency_in_literacy": proficiency_in_literacy,
                    "proficiency_in_numeracy": proficiency_in_numeracy,
                    "employment_status": employment_status,
                    "below_poverty_line": below_poverty_line,
                    "time_spent_on_unpaid_domestic_work": time_spent,
                    "survey_year": year,
                    "takes_by_personal_id": personal_id
                }
                surveys.append(survey)

                personal_id += 1
    return individuals, surveys

# Step 4: Export data to CSV files
def export_to_csv(filename, data, fieldnames):
    """
    Export a list of dictionaries 'data' to a CSV file with given 'fieldnames'.
    """
    with open(filename, 'w', newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        for row in data:
            writer.writerow(row)

def main():
    # 4.1 Load regions from CSV (ensure regions.csv is present)
    regions = load_regions('regions.csv')

    # 4.2 Generate yearly parameters for survey indicators
    yearly_params = generate_yearly_params()

    # 4.3 Generate INDIVIDUAL and SURVEY data
    # sample_size=5000 means each region-year will have 5000 records
    individuals, surveys = generate_data(regions, yearly_params, sample_size=5000)

    # 4.4 Export INDIVIDUAL data to CSV
    export_to_csv('individuals.csv', individuals, ['personal_id', 'sex', 'age', 'lives_in_region_id'])

    # 4.5 Export SURVEY data to CSV
    export_to_csv('surveys.csv', surveys, [
        'survey_id',
        'proficiency_in_literacy',
        'proficiency_in_numeracy',
        'employment_status',
        'below_poverty_line',
        'time_spent_on_unpaid_domestic_work',
        'survey_year',
        'takes_by_personal_id'
    ])

    print("Data generation complete. 'individuals.csv' and 'surveys.csv' have been created.")

if __name__ == "__main__":
    main()
```
